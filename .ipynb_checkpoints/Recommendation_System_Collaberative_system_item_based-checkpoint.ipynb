```python
# To make recommendation system,required variables are
# genres,keywords,overview, title, cast, crew, movie_id
```

```python
# Selecting only the important string variables for recommendation system

movies = movies[['movie_id','title','genres','keywords','overview',  'cast', 'crew']]
```

```python
# Reducing the number of variables to 3
# Since genres, keywords, cast and crew are in dictionary format, we use
# ast module to extraxt value of the key specified and return list
# We only look for "name" key in genres

import ast

def convert(text):
    lst = []
    for i in ast.literal_eval(text):
        lst.append(i['name'])
    return lst
```

```python
# function for cast to extract top 3 actors

def convert3(text):
    lst = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter<3:
            lst.append(i['name'])
            counter+=1
    return lst
```

```python
# function to extract director name from crew variable

def director(text):
    lst = []
    for i in ast.literal_eval(text):
        if i['job']== 'Director':
            lst.append(i['name'])
    return lst
```

```python
#Applying convert function in genres variable 

movies['genres'] = movies['genres'].apply(convert)
```

```python
# Applying same function to keywords variable

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# Applying convert3 function to cast variable

movies['cast'] = movies['cast'].apply(convert3)
```

```python
# Applying director fun on crew variable

movies['crew'] = movies['crew'].apply(director)
```

```python
movies.head(2)
```

```python
# Remove the space bt first and last names for using NLP
# Reducing feature size

def collapse(L):
    lst = []
    for i in L:
        lst.append(i.replace(" ",""))
    return lst
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
```

```python
movies.head(2)
```

```python
# Missing values in overview variable
movies.isnull().sum()
```

```python
# Dropping na values

movies.dropna(inplace=True)
```

```python
# Tokenizing overview variable

movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

```python
# Appending the variables

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] +movies['crew']
```

```python
movies.drop(columns= ['genres','keywords','overview','cast','crew'],axis=1,inplace=True)
```

```python
movies.head(2)
```

```python
movies.info()
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

## NLP  - Bag of words

```python
len(movies['tags'])
```

```python
# Using count vectorizer to create the vectors for each term

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Creating count vectorizer object by removing english stop words and setting max_features
# features(unique terms) from the training set can be more but we are considering ly top 5000
# high weightage terms
# Since machine can not handle texts, using CV to convert to vector values

cv = CountVectorizer(stop_words= 'english',max_features=5000,binary=True )
```

```python
# Listing first 100 features 

cv.get_feature_names()[:100]
```

```python
# Fit and transforming the object with dataset and converting into array

vector = cv.fit_transform(movies['tags']).toarray()
```

```python
# Vocabulary is the collection of all the features 
# Total documnets is 4806 and we have selected max features as 5000, hence 5000 vector values for 
# each documnet, values are not-zero if the term is present in the vocabulary else 0

vector.shape
```

```python
# Apply Cosine Similarity on the vectors since the data is high
# Cosine Similarity - Angle between 2 vectors
# high angle - less similarity
# less angle - high similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Finding Cosine similarity of each vector to other vectors 

similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
# Similarity scores between all the movies in array
# Sort the scores and the movie with high score is similar to that respective movie 

similarity
```

```python
# Accessing index value of each movie by using the code below

movies[movies['title']=='Avatar'].index[0]
```

```python
# Similarity values of first movie in the list
similarity[0]
```

```python
# creating function to find the similar movies
# Similar movies retrieved based on the index values

def recommend(movie):
    index = movies[movies['title']==movie].index[0] #finding index value
    similar = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x:x[1])  
    
    for i in similar[1:6]: #finding first 5 similar movies
        print(movies.iloc[i[0]].title)
        
    
```

```python
recommend('Batman')
```
